## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# establish bins and group names
bins = [0, 10, 15, 20, 25, 30, 35, 40, 100]
age_grps = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

# categories players into age bins
purchase_data["Age Range"] = pd.cut(purchase_data["Age"], bins, labels=age_grps, right=False)
purchase_data

# player count by age group
# get list of unique players using updated purchase_data records
players = purchase_data.drop_duplicates('SN')
age_ct = players.groupby('Age Range')['SN'].count()

# Calculate percentages by age group
age_range_pct = (age_ct/total_players)*100

# create new dataframe with to display formatted data
age_demographics = pd.DataFrame({"Total Count": age_ct,
                                 "Percentage of Players": age_range_pct})
age_demographics["Percentage of Players"]=age_demographics["Percentage of Players"].map("{0:,.2f}%".format)

#remove "age range" from index name
age_demographics.index.name = None

age_demographics
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python

pop_table = purchase_data[["Item ID","Item Name","Price"]]

purch_count = pop_table.groupby("Item ID")["Item Name"].count()
purch_count

item_name = pop_table["Item Name"]

item_price = pop_table["Price"]
item_price

purch_value = (item_price * purch_count).map("${:.2f}".format)
purch_value

most_pop = pd.DataFrame({"Item Name": item_name,
                         "Purchase Count": purch_count,
                        "Price per Item": item_price,
                        "Total Purchase Value": purch_value})

most_pop = most_pop.sort_values("Purchase Count", ascending=False)

most_pop["Price per Item"] = most_pop["Price per Item"].map("${:.2f}".format)

most_pop = most_pop.set_index("Item Name")

most_pop.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order

```python
purch_value = (item_price * purch_count)

most_profit = pd.DataFrame({"Item Name": item_name,
                         "Purchase Count": purch_count,
                        "Price per Item": item_price,
                        "Total Purchase Value": purch_value})

most_profit = most_profit.sort_values("Total Purchase Value", ascending=False)
  
most_profit["Price per Item"] = most_profit["Price per Item"].map("${:.2f}".format)
most_profit["Total Purchase Value"] = most_profit["Total Purchase Value"].map("${:.2f}".format)

most_profit.head()
```
